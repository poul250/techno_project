```python
cat_features = pd.DataFrame(data[['person_status']])
cat_features = pd.DataFrame([str(j).lower() for j in cat_features.values])
label_encoder = LabelEncoder()
df = pd.DataFrame(label_encoder.fit_transform(cat_features))
data = data.drop('person_status', axis=1)
df_onehotencoder = OneHotEncoder(sparse=False)
df2 = df_onehotencoder.fit_transform(df)
df = pd.DataFrame(df2)
data = data.join(df, rsuffix="_pers")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
len(data.columns)
data=data.dropna(axis=0)
```

```python
data.columns
```

```python
arr = np.array(data)
```

```python
arr.shape
arr = arr[:, 1:]
arr[:, 1] = arr[:, 1] - arr[:, 0]
arr = arr[:, 1:]
```

```python
scaler = StandardScaler()
scaler.fit(arr)
arr=scaler.transform(arr)
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=8)
```

```python
clf.fit(arr[:, 1:])
```

```python
size = []
for i in range(8):
    size += [((i == clf.labels_).sum())]
    print(size[i])
```

```python
for i in range(len(clf.cluster_centers_)):
    print(i, clf.cluster_centers_[i])
```

```python
data.loc[:, 'labels'] = clf.labels_
clusters = data.groupby('labels').mean()
clusters
for k in range(8):
    print(k, clusters.loc[k])
    print('size = ', size[k])
```
